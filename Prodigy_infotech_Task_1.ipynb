```python
import pandas as pd
import numpy as np


from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

```

```python
folder_path= '/content/drive/MyDrive/New folder'
```

```python
train_data= pd.read_csv(folder_path+"/train.csv")
test_data=pd.read_csv(folder_path+"/test.csv")
```

```python
x= train_data[['LotArea','TotRmsAbvGrd','OverallCond']]
```

```python
y=train_data['SalePrice']
```

```python
x.describe()
```

```python
regressor.fit(x,y)
```

```python
x_pred=regressor.predict(x)
```

```python
x_test= test_data[['LotArea','TotRmsAbvGrd','OverallCond']]
```

```python
test_data.columns
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae= mean_absolute_error(x_pred,train_data['SalePrice'])
mse= mean_squared_error(x_pred,train_data['SalePrice'])
r2= r2_score(x_pred,train_data['SalePrice'])
print("Mean Absolute Error is:",mae)
print("Mean Squared Error is:",mse)
print("R2 Score is:",r2)
```
